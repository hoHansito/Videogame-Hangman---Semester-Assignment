```python
import random
import time
from IPython.display import clear_output

def clear(): #REINICIO DEL OUTPUT
    clear_output(wait=False)

def diccionario(): #LISTA DE PALABRAS INGRESADAS
    lp = []
    while True:
        p = str(input("Ingresa una palabra para agregar al juego: "))
        if p != "":
            lp.append(p)
        else:
            return lp

def timmy2(errores): #PRINT DEl AHORCADO
    vidas1 = str(errores)
    vidas2 = errores + 1
    vidas2 = str(vidas2)
    with open('Timmy.txt', 'r', encoding='utf-8') as archivo:
        imprime = False
        for linea in archivo:
            if vidas1 in linea:
                imprime = True
            elif vidas2 in linea:
                break
            elif imprime:
                print(linea.rstrip())

def defeat_screen(word,name): #PRINT DE PANTALLA DE DERROTA
    clear()
    with open('DEFEAT.txt', 'r', encoding='utf-8') as archivo:
        imprime = False
        for linea in archivo:
            if "inicio" in linea:
                imprime = True
            elif "dead" in linea:
                print(f"{name} ha sido ahorcado...")
            elif "word" in linea:
                print(word)
            elif "final" in linea:
                break
            elif imprime:
                print(linea.rstrip())
            time.sleep(0.25)

def victory_screen(word): #PRINT DE PANTALLA DE VICTORIA
    clear()
    with open('VICTORY.txt', 'r', encoding='utf-8') as archivo:
        imprime = False
        for linea in archivo:
            if "inicio" in linea:
                imprime = True
            elif "word" in linea:
                print(word)
            elif "final" in linea:
                break
            elif imprime:
                print(linea.rstrip())
            time.sleep(0.25)

def select_word(lp): #LETRAS INSERTADAS EN LOS ESPACIOS Y SELECTOR DE PALABRA
    word = random.choice(lp)
    word = word.upper()
    wordl = []
    wordlt = []
    wordsel = []
    vtilde = ["Á","É","Í","Ó","Ú"]
    vsintilde = ["A","E","I","O","U"]
    for char in word:
        if char in vtilde:
            pos = vtilde.index(char)
            sintilde = vsintilde[pos]
            wordl.append(sintilde)
            wordlt.append(char)
            wordsel.append(False)
        elif char != "\n":
            wordl.append(char)
            wordlt.append(char)
            wordsel.append(False)
    return word,wordl,wordlt,wordsel

def isletter(l,wordl,wordsel,letrasusadas): #ANALISIS DE LA LETRA INSERTADA
    pos = 0
    letracorrecta = False
    for elem in wordl:
        if elem == l:
            wordsel.pop(pos)
            wordsel.insert(pos,True)
            letracorrecta = True
        pos += 1
    return l,wordsel,letracorrecta

def palabra_(wordlt,wordsel): #ESPACIOS DE LA PALABRA ALEATORIA
    print("")
    print("Palabra: ")
    print(" ")
    pos = 0
    word_ = ""
    for i in wordsel:
        if i:
            letra = wordlt[pos]
            word_ = word_ + " " + letra
        else:
            word_ = word_ + " _"
        pos += 1
    return word_

def leeValida(): #VALIDACIÓN DE LETRAS INSERTADAS
    while True:
        l = input()
        if l.isalpha() and len(l) == 1:
            return l
        else:
            print("Error, caracter no válido.")

def jugar(): #JUEGO PRINCIPAL
    playing = True
    reset_list = True
    while playing == True:
        if reset_list:
            lp = diccionario()
            reset_list = False
        else:
            name = str(input("Introduzca el nombre de su Ahorcado: "))
            errores = 0
            letrasusadas = []
            word,wordl,wordlt,wordsel = select_word(lp)
            while False in wordsel:
                clear()
                timmy2(errores)
                print(f"Letras Usadas: {letrasusadas}")
                print(f"Errores: {errores}/10")
                print(palabra_(wordlt,wordsel))
                print(" ")
                l = leeValida()
                l = l.upper()
                if l not in letrasusadas:
                    letrasusadas.append(l)
                l,wordsel,letracorrecta = isletter(l,wordl,wordsel,letrasusadas)
                if not letracorrecta:
                    errores += 1
                if errores == 10:
                    clear()
                    timmy2(errores)
                    time.sleep(1.5)
                    defeat_screen(word,name)
                    break
                letrasusadas.sort()
            while all(wordsel):
                victory_screen(word)
                break
            n = int(input(""))
            if n == 1:
                reset_list = False
                pass
            if n == 2:
                reset_list = True
            if n == 3:
                playing = False
```

```python
jugar()
```
